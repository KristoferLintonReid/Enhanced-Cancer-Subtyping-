```r
Z <- merge(clincal, groups, by='X')
```

```r
tail(Z)
```

```r
library(varhandle)
Z$Survival_months <- unfactor(Z$Survival_months)
```

```r
library(survival)                # for Surv(), survfit()
library(survminer)
```

```r
nrow(Z)
```

```r
str(Z)
```

```r
KM <- survfit(Surv(as.numeric(Survival_months), Status) ~ cluster , type="kaplan-meier", conf.type="log", data=Z)
```

```r
cox <- coxph(Surv(Z$Survival_months, Z$Status) ~ Z$cluster + Z$Stage)
summary(cox)
```

```r
Cox_fit <- survfit(cox)
autoplot(Cox_fit)
```
