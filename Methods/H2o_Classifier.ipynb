```r
library(h2o)
```

```r
setwd('media')
```

```r
y <- read.csv('order_Classes.csv', row.names=1)
```

```r
x <- read.csv('order_Normdat.csv', row.names=1)
```

```r
x<- as.data.frame(x)
```

```r
df <- merge(x,y, by='row.names')
rownames(df) <- df$Row.names
df$Row.names <- NULL 
```

```r
tail(df)
```

```r
n <- nrow(df)
train_i <- 1:round(0.7*n)
test_i <- round(0.7*n+1):n
train <- df[train_i,]
test <- df[test_i,]
```

```r
h2o.init(nthreads=-1)
```

```r
test <-  as.h2o(test)
```

```r
train <-  as.h2o(train)
```

```r
h2o.tail(train)
```

```r
h2o.describe(train)
```

```r
var_genes <- apply(t(df), 1, var)
head(var_genes)
```

```r
select_var <- names(sort(var_genes, decreasing=TRUE))[1:500]
head(select_var)
```

```r
y <- "X"
x <- select_var
```

```r
str(x)
str(y)
```

```r
aml <- h2o.automl(x = x, y = y,
                  training_frame = train,
                  max_models = 100,
                  seed = 1)
```

```r
lb <- aml@leaderboard
```

```r
str(lb)
```

```r
d <- head(lb, n=10)
d
```

```r
write.csv(d, file= 'top_10_models.csv')
```

```r
nrow(lb)
```

```r
str(lb)
```

```r
# Get model ids for all models in the AutoML Leaderboard
model_ids <- as.data.frame(aml@leaderboard$model_id)[,1]
# Get the "All Models" Stacked Ensemble model
se <- h2o.getModel(grep("StackedEnsemble_AllModels", model_ids, value = TRUE)[1])
# Get the Stacked Ensemble metalearner model
metalearner <- h2o.getModel(se@model$metalearner$name)
```

```r
h2o.varimp(metalearner)
```

```r
h2o.varimp_plot(metalearner)
```

```r
sum_al <- summary(aml@leader)
```

```r
sum_al
```

```r
perf <- h2o.performance(aml@leader, test)
h2o.confusionMatrix(perf)
conf_mat <- h2o.confusionMatrix(perf)
write.csv(conf_mat, file='conf_mat.csv')
```
